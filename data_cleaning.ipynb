```python
#Dependecies
import pandas as pd
import os 
```

```python
#Xbox 1 csv
file = 'Resources/XboxOne_GameSales.csv'
xbox = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
#Ps4 csv
file2 = 'Resources/PS4_GamesSales.csv'
ps4 = pd.read_csv(file2, encoding="ISO-8859-1")
```

```python
xbox.head(1)
```

```python
ps4.head()
```

```python
#cleaning 
del xbox["Pos"]
```

```python
#Merge
console_wars= pd.merge(xbox,ps4, on="Game", suffixes= ('_xbox', '_ps4'))
console_wars.head()
```

```python
console_wars= console_wars.dropna(how= 'any')
```

```python
console_wars['Year_xbox']= console_wars['Year_xbox'].astype(int) 
```

```python
console_wars['Year_ps4']= console_wars['Year_ps4'].astype(int) 
```

```python
console_wars['Total_Global_Sales']= console_wars['Global_xbox']+ console_wars['Global_ps4']
```

```python
del console_wars["Year_ps4"]
```

```python
del console_wars["Genre_ps4"]
```

```python
del console_wars["Publisher_ps4"]
```

```python
console_wars= console_wars.rename(columns= {'Year_xbox':'Year', 'Genre_xbox': 'Genre', 'Publisher_xbox':'Publisher', "North America_xbox": "Xbox_NA_Sales", "Europe_xbox": "Xbox_EU_Sales", "Japan_xbox": "Xbox_JP_Sales", "Rest of World_xbox": "Xbox_RoW_Sales", "Global_xbox": "Xbox_Global_Sales","North America_ps4": "PS4_NA_Sales", "Europe_ps4": "PS4_EU_Sales", "Japan_ps4": "PS4_JP_Sales", "Rest of World_ps4": "PS4_RoW_Sales", "Global_ps4": "PS4_Global_Sales" })
console_wars.head()
```

```python
Games= console_wars[['Game', 'Year', 'Genre', 'Publisher', 'Total_Global_Sales']]
Games["Game_ID"] = Games.index + 1
Games.to_csv("data/Games.csv", header=True, index=False)
```

```python
xbox_games= console_wars[['Xbox_NA_Sales', 'Xbox_EU_Sales', 'Xbox_JP_Sales', 'Xbox_RoW_Sales', 'Xbox_Global_Sales']]
xbox_games["Game_ID"] = Games["Game_ID"]
xbox_games.to_csv("data/xbox_games.csv", header=True, index=False)
```

```python
ps4_games= console_wars[['PS4_NA_Sales', 'PS4_EU_Sales', 'PS4_JP_Sales', 'PS4_RoW_Sales', 'PS4_Global_Sales']]
ps4_games["Game_ID"] = Games["Game_ID"]
ps4_games.to_csv("data/ps4_games.csv", header=True, index=False)
```
